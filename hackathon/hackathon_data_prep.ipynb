source: https://collegescorecard.ed.gov/data

This file was used to create the files in the `data\` folder. So run, download and unzip the files from the College Scorecard website and save them into a folder called `college_scorecard`. Beware, as writting, this code loads the entirety of the college scorecard data into memory and the total filesize is very large. This was completed on a machine with 64GB of ram. At least 32GB of ram is recommended to follow this notebook. This can be done on a smaller machine, but you will need to rewrite code below to process data in smaller batches.

```python
# Create field_map.csv from the data dictionary yaml file. This file is the data dictionary for the College Scorecard dataset.
with open('./college_scorecard/College_Scorecard_Raw_Data_01162025/data.yaml', 'r') as file:
    data_dictionary = yaml.safe_load(file)

field_map = pd.DataFrame(data_dictionary['dictionary']).T
field_map.to_csv('./data/field_map.csv', index=False)
```

```python
#load all the csv files in the directory into a dictionary of dataframes. 
csv_files = [file for file in os.listdir('college_scorecard/College_Scorecard_Raw_Data_01162025/') if file.endswith('.csv')]
print(csv_files)
data_frames = {file: pd.read_csv(f'college_scorecard/College_Scorecard_Raw_Data_01162025/{file}', low_memory=False) for file in csv_files}
```

```python
# create a new dataframe that contains all the field of study dataframes concatenated together.
field_of_study_frames = [frame for frame in data_frames.keys() if frame.startswith('FieldOfStudyData')]

field_study_all_years = pd.concat(
    [data_frames[frame].assign(file_name=frame) for frame in field_of_study_frames],
    ignore_index=True
)
```

```python
#create a dataframe to only include rows where the INSTNM is Carthage College.
field_study_all_years_carthage = field_study_all_years[field_study_all_years.INSTNM=='Carthage College']
```

```python
#save the field_study_all_years and field_study_all_years_carthage dataframes to csv files.
field_study_all_years.to_csv('./data/field_study_all_years.csv', index=False)
field_study_all_years_carthage.to_csv('./data/field_study_all_years_carthage.csv', index=False)
```

```python
# create a new dataframe that contains all the institution dataframes concatenated together.
institution_frames = [frame for frame in data_frames.keys() if frame.startswith('MERGED')]

institution_all_years = pd.concat(
    [data_frames[frame].assign(file_name=frame) for frame in institution_frames],
    ignore_index=True
)
```

```python
#create a dataframe to only include rows where the INSTNM is Carthage College.
institution_all_years_carthage = institution_all_years[institution_all_years.INSTNM=='Carthage College']
```

```python
#save the institution_all_years and institution_all_years_carthage dataframes to csv files.
institution_all_years.to_csv('./data/institution_all_years.csv', index=False)
institution_all_years_carthage.to_csv('./data/institution_all_years_carthage.csv', index=False)
```
